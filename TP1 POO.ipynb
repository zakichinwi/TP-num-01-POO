```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_correlation_map(df):
    numeric_columns = df.select_dtypes(include=['number'])#select_dtypes() sélectionne uniquement les colonnes numériques du DataFrame df et les stocke dans la variable numeric_columns
    corr = numeric_columns.corr() 
    plt.figure(figsize=(12, 10))#définit la taille de la figure à 12 pouces de large et 10 pouces de haut
    cmap = sns.diverging_palette(220, 10, as_cmap=True) 
    sns.heatmap(corr, cmap=cmap, square=True, cbar_kws={'shrink': 0.9}, annot=True, annot_kws={'fontsize': 12})
    
    plt.show()#affichage de heatmap
plot_correlation_map(df)#cette ligne du code génére un heatmap montrant les relations de corrélation entre les différentes colonnes numériques du DataFrame on remarque une relation de lineariter
```
